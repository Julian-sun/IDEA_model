```python
import pandas as pd
import numpy as np
from scipy.stats import nbinom
from pathlib import Path


import matplotlib.pyplot as plt
```

```python
def calculate_expected_cases_nb(df, cases = 'atend_ivas', min_years=2, fallback_dispersion=1000):
    """
    Estimate expected cases per epiweek using historical data and Negative Binomial model.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Must contain columns: ['year', 'epiweek', 'cases']
    min_years : int
        Minimum number of past years needed to estimate NB parameters. Otherwise, uses Poisson approx.
    fallback_dispersion : float
        Dispersion (r) value to use if variance <= mean (e.g., near-Poisson case)

    Returns:
    --------
    pd.DataFrame
        Columns: ['epiweek', 'expected_cases', 'ci_lower_95', 'ci_upper_95', 'n_years']
    """
    results = []

    for week in range(1, 54):  # Support epiweek 53 in some years
        weekly_data = df[df['epiweek'] == week][cases].dropna()

        n_years = weekly_data.shape[0]
        if n_years < min_years:
            continue  # Skip if not enough data

        mu = weekly_data.mean()
        var = weekly_data.var(ddof=1)

        # Estimate dispersion parameter r
        if var > mu:
            r = mu**2 / (var - mu)
        else:
            r = fallback_dispersion  # Use large r to approximate Poisson

        p = r / (r + mu)

        # 95% Confidence Interval
        ci_lower = nbinom.ppf(0.025, r, p)
        ci_upper = nbinom.ppf(0.975, r, p)

        results.append({
            'epiweek': week,
            'expected_cases': mu,
            'ci_lower_95': ci_lower,
            'ci_upper_95': ci_upper,
            'n_years': n_years
        })

    return pd.DataFrame(results)
```

```python
mun_new = list(Path('/opt/storage/refined/aesop/visualization/')
               .glob('aesop_*_mun_new.parquet'))

aesop_mun_new = max(mun_new, key=lambda x: x.stat().st_ctime)

df = pd.read_parquet(aesop_mun_new)
```

```python
code = 291875
set_muni_2022_2024 = df[(df.co_ibge == code) & (df.epiyear >= 2022) & (df.epiyear <= 2024)]
set_muni_2025 = df[(df.co_ibge == code) & (df.epiyear == 2025)]
set_muni_2025 = set_muni_2025[['co_ibge','epiyear','epiweek','atend_ivas','sinal_ens_ivas']]
```

```python
set_muni_2022_2024
```

```python
set_muni_2022_2024 = set_muni_2022_2024[['co_ibge','epiyear','epiweek','atend_ivas','num_otc_ivas','sinal_ens_ivas']]
```

```python
expected_df = calculate_expected_cases_nb(set_muni_2022_2024)
```

```python
expected_df
```

```python
set_muni_2025['epiweek'] = set_muni_2025['epiweek'].astype(str)
expected_df ['epiweek'] = expected_df ['epiweek'].astype(str)

merged = set_muni_2025.merge(expected_df, on=['epiweek'], how='left')

```

```python
merged[['epiweek', 'atend_ivas', 'expected_cases',
       'ci_lower_95', 'ci_upper_95','sinal_ens_ivas']]
```

```python
# Plot
plt.figure(figsize=(14, 6))

# Confidence interval
plt.fill_between(merged['epiweek'], merged['ci_lower_95'], merged['ci_upper_95'],
                 color='lightblue', alpha=0.5, label='95% CI (NB)')

# Expected cases
plt.plot(merged['epiweek'], merged['expected_cases'], color='blue', label='Expected cases (mean)')

# Observed cases
plt.plot(merged['epiweek'], merged['atend_ivas'], color='black', marker='o', label='Observed cases (atend_ivas)')

# Highlight points outside CI
outside_ci = (merged['atend_ivas'] < merged['ci_lower_95']) | (merged['atend_ivas'] > merged['ci_upper_95'])
plt.scatter(merged.loc[outside_ci, 'epiweek'],
            merged.loc[outside_ci, 'atend_ivas'],
            color='orange', label='Outside 95% CI', zorder=5)

# Orange stars for sinal_aesop_ivas == 1
warning_weeks = merged[merged['sinal_ens_ivas'] == 1]
plt.scatter(warning_weeks['epiweek'], warning_weeks['atend_ivas'],
            color='red', marker='*', s=180, label='AESOP warning (sinal_ens_ivas)') #edgecolor='darkred'

#plt.ylim([0, 10000])
# Labels
plt.xlabel('Epidemiological Week')
plt.ylabel('Number of Cases')
plt.title('Observed vs Expected Weekly Cases with 95% CI (Negative Binomial) - 2025')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
